```python
# pick window on each side: k=6 ≈ 1 day each side on 4h bars
k = 6
roll_max = df['perp_close'].rolling(2*k+1, center=True).max()
roll_min = df['perp_close'].rolling(2*k+1, center=True).min()
df['is_top']    = df['perp_close'].eq(roll_max)
df['is_bottom'] = df['perp_close'].eq(roll_min)

print("tops:", int(df['is_top'].sum()), "bottoms:", int(df['is_bottom'].sum()))

# quick plot with markers
import matplotlib.pyplot as plt
tops    = df.index[df['is_top']]
bottoms = df.index[df['is_bottom']]

fig, ax = plt.subplots(figsize=(12,5))
ax.plot(df.index, df['perp_close'], lw=1.2, color='tab:blue', label='Perp')

ax.vlines(tops,    ymin=df['perp_close'].min(), ymax=df['perp_close'].max(),
          colors='crimson', alpha=0.15, lw=1)
ax.vlines(bottoms, ymin=df['perp_close'].min(), ymax=df['perp_close'].max(),
          colors='green',   alpha=0.15, lw=1)

ax2 = ax.twinx()
ax2.step(df.index, df['funding_rate'], where='post', lw=1.0, color='tab:orange', label='Funding')

ax.set_title('4H price with centered tops/bottoms; funding step')
ax.legend(loc='upper left')
plt.show()
```

```python
import numpy as np

sign_now  = np.sign(df['funding_rate'])
sign_prev = np.sign(df['funding_rate'].shift(1))

# flip if sign changes and neither side is exactly zero
df['flip'] = sign_now.ne(sign_prev) & sign_now.ne(0) & sign_prev.ne(0)

# optional: direction-specific flags
df['flip_pos2neg'] = (df['funding_rate'] < 0) & (df['funding_rate'].shift(1) > 0)
df['flip_neg2pos'] = (df['funding_rate'] > 0) & (df['funding_rate'].shift(1) < 0)

# plot flips on price
flips = df.index[df['flip']]

fig, ax = plt.subplots(figsize=(12,5))
ax.plot(df.index, df['perp_close'], lw=1.2, color='tab:blue')
ax.scatter(flips, df.loc[flips,'perp_close'], s=18, marker='^', color='black', label='Funding flip', zorder=3)

ax2 = ax.twinx()
ax2.step(df.index, df['funding_rate'], where='post', lw=1.0, color='tab:orange')

ax.set_title('Funding sign flips (▲) overlaid on price')
ax.legend(loc='upper left')
plt.show()
```

```python
eps = 1e-8

f_now  = df['funding_rate']
f_prev = df['funding_rate'].shift(1)

# + → −
df['flip_pos2neg'] = (f_prev >  +eps) & (f_now <  -eps)

# − → +
df['flip_neg2pos'] = (f_prev <  -eps) & (f_now >  +eps)

# if you still want a combined flag:
df['flip_any'] = df['flip_pos2neg'] | df['flip_neg2pos']

p2n = df.index[df['flip_pos2neg']]
n2p = df.index[df['flip_neg2pos']]

ax.plot(df.index, df['perp_close'], lw=1.2, color='tab:blue')
ax.scatter(p2n, df.loc[p2n,'perp_close'], s=18, c='crimson', marker='v', label='+→−', zorder=3)
ax.scatter(n2p, df.loc[n2p,'perp_close'], s=18, c='limegreen', marker='^', label='−→+', zorder=3)
ax.legend(loc='upper left')

print("pos→neg:", int(df['flip_pos2neg'].sum()))
print("neg→pos:", int(df['flip_neg2pos'].sum()))
plt.show()
```

```python
#distibution chart

x = df['funding_rate'].dropna().values
p25, p50, p75 = np.percentile(x, [25,50,75])
mean = np.mean(x)

fig, ax = plt.subplots(figsize=(8,4))
ax.hist(x, bins=50, alpha=0.8)
ax.axvline(mean,  color='tab:blue',  linestyle='--', lw=1.2, label=f'mean {mean:.2e}')
ax.axvline(p50,   color='tab:orange',linestyle='--', lw=1.2, label=f'median {p50:.2e}')
ax.axvline(p25,   color='tab:green', linestyle=':',  lw=1.0, label=f'P25 {p25:.2e}')
ax.axvline(p75,   color='tab:red',   linestyle=':',  lw=1.0, label=f'P75 {p75:.2e}')
ax.set_title('Funding rate — distribution (all bars, no time)')
ax.set_xlabel('Funding rate')
ax.set_ylabel('Count')
ax.legend()
ax.grid(True, alpha=0.15)
plt.show()
```

```python
import math

ranges = [
    ("2025-05-20", "2025-05-24"),
    ("2025-06-08", "2025-06-12"),
    ("2025-07-12", "2025-07-16"),
    ("2025-08-12", "2025-08-16"),
]

rows = math.ceil(len(ranges) / 2)
cols = 2

fig, axes = plt.subplots(rows, cols, figsize=(14, 3.2*rows), squeeze=False)
axes = axes.flatten()

for i, (start, end) in enumerate(ranges):
    ax1 = axes[i]
    w = df.loc[start:end, [PRICE_COL, FUND_COL]].copy()

    ax1.plot(w.index, w[PRICE_COL], linewidth=1.4)
    ax1.set_ylabel("Perp Close")

    ax2 = ax1.twinx()
    ax2.plot(w.index, w[FUND_COL], linestyle="--", linewidth=1.0, color="orange")
    ax2.set_ylabel("Funding")

    ax1.set_title(f"{start} → {end}")
    ax1.grid(True, axis="y", alpha=0.25)

# hide unused axes
for j in range(i+1, rows*cols):
    axes[j].axis("off")

fig.suptitle("Perp vs Funding (multiple windows)", y=0.995)
fig.tight_layout()
plt.show()
```
